```python
import numpy as np
import pandas as pd
from sklearn import datasets
from sklearn.ensemble import RandomForestClassifier
iris=datasets.load_iris()
print("data :",iris.data)
print("target :",iris.feature_names)
```

```python
df=pd.DataFrame(iris.data,columns=iris.feature_names)
print(df)
```

```python
#now adding output row
df["species"]=pd.Categorical.from_codes(iris.target,iris.target_names)
df.head()
```

```python
df["is_train"]=np.random.uniform(0,1,len(df))<=0.75
df
```

```python
train=df[df["is_train"]==True]
test=df[df["is_train"]==False]
print("lengths :",len(train))
print("lengths :",len(test))
```

```python
y=df.columns[0:4]
k=pd.factorize(train["species"])[0]
k1=pd.factorize(test["species"])[0]
print(k1)
```

```python
model=RandomForestClassifier(n_estimators=100,n_jobs=2,random_state=2)
```

```python
model.fit(train[y],k)
```

```python
q=model.predict(test[y])
```

```python
c=0
for i in range(0,len(q)):
    if(q[i]==k1[i]):
        c=c+1
per=(c/len(q))*100
print(per)
```
